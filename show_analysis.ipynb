```python
## Importing the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Getting the data
tv_data = pd.read_csv("C:/Users/Ryanm/DataScienceProjects/streaming-platform-analysis/tv_shows.csv")
tv_data = tv_data.drop(['type'], axis=1)
tv_data.head()
```

```python
## Creating variables for each individual streaming platform
netflix = tv_data.loc[tv_data['Netflix'] == 1]
hulu = tv_data.loc[tv_data['Hulu'] == 1]
prime_video = tv_data.loc[tv_data['Prime Video'] == 1]
disney = tv_data.loc[tv_data['Disney+'] == 1]
```

```python
## Pie chart comparing the number of shows for each streaming service
label=['Netflix', 'Hulu', 'Prime Video', 'Disney+']
data=[len(netflix.index), len(hulu.index), len(prime_video.index), len(disney.index)]
explode = (0, 0, 0.1, 0)

plt.pie(data, labels=label, explode=explode, autopct='%1.1f%%')
plt.title('Number of TV Shows by Streaming Service')
plt.axis('equal')
plt.show()
```

```python
label=['13+', '16+', '18+', '7+', 'all']
plt.pie(netflix.groupby('Age').size(), labels=label)

plt.title('Proportion of TV Shows on Netflix by Age Approporiateness')
plt.axis('equal')
plt.show()
```

```python
label=['16+', '18+', '7+', 'all']
plt.pie(hulu.groupby('Age').size(), labels=label)

plt.title('Proportion of TV Shows on Hulu by Age Approporiateness')
plt.axis('equal')
plt.show()
```

```python
label=['13+', '16+', '18+', '7+', 'all']
plt.pie(prime_video.groupby('Age').size(), labels=label)

plt.title('Proportion of TV Shows on Prime Video by Age Approporiateness')
plt.axis('equal')
plt.show()
```

```python
label=['16+', '7+', 'all']
plt.pie(disney.groupby('Age').size(), labels=label)

plt.title('Proportion of TV Shows on Prime Video by Age Approporiateness')
plt.axis('equal')
plt.show()
```
